# Introduction

In this project we will be examining a dataset which looks at the gender gap between college majors. We will be creating visualizations of the data in matplotlib, make the charts look clean and easy to read, and we will discuss the results. Let's start by examining the gender gap in STEM majors.

# STEM majors gender gap

```python
#Import the matplotlib and pandas libraries
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

#Read in our dataset, and assign to the variable, women_degrees
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

#Create a list of stem majors
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

#Create a figure
fig = plt.figure(figsize=(18, 3))

#Generate 6 charts for each of the STEM majors
for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_orange, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")

#Create labels for 'Men' and 'Women' in the charts    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        
#Display the charts
plt.show()
```

The charts shown above are all from STEM majors, and ordered by decreasing gender gap. We can see that the greatest gender gap is in the Engineering major, with there being more male engineering majors than female. The smallest gap is in Math and Statistics. In Psychology and Biology, there are more female than male graduates. 

Perhaps the results for the Biology major are the most interesting, because prior to 1985 there were more male graduates. But after 1985 the trend switched, and now there are more female Biology graduates.

# Comparing across all degrees

While the previous charts we generated were specific to STEM majors, we want to examine the gender gap across all majors. Now let's create charts for all the majors.

```python
#List of STEM majors
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
#List of liberal arts majors
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
#List of other majors
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

#Create a plotting figure with the dimensions 16X20
fig = plt.figure(figsize=(16, 20))

#Create charts for the stem_cats degrees
for sp in range(0, 18, 3):                #sp = [0, 3, 6, 9, 12, 15]
    cat_index = int(sp/3)                 #cat_index = [0, 1, 2, 3, 4, 5]. This line indexes through stem_cats
    ax = fig.add_subplot(6,3,sp+1)        #Subplot positions = [1, 4, 7, 10, 13, 16]
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_index]], c=cb_orange, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[cat_index]], c=cb_dark_blue, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[cat_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    #Create labels for 'Men' and 'Women' in the charts    

    if cat_index == 0:
        ax.text(2005, 87, 'Women')
        ax.text(2005, 12, 'Men')
    elif cat_index == 5:
        ax.text(2005, 90, 'Men')
        ax.text(2005, 5, 'Women')
        

#Create charts for the liberal arts degrees
for sp in range(1, 16, 3):                     #sp = [1, 4, 7, 10, 13]
    cat_index = int((sp-1) / 3)                #cat_index = [0, 1, 2, 3, 4]
    ax = fig.add_subplot(6,3,sp+1)             #Subplot positions = [2, 5, 8, 11, 14, 17]
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_index]], c=cb_orange, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[cat_index]], c=cb_dark_blue, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[cat_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    #Create labels for 'Men' and 'Women' in the charts    

    if cat_index == 0:
        ax.text(2005, 80, 'Women')
        ax.text(2005, 20, 'Men')
            
       
#Create charts for the other degrees
for sp in range(2, 20, 3):                      #sp = [2, 5, 8, 11, 14, 17]
    cat_index = int((sp-2) / 3)                 #cat_index = [0, 1, 2, 3, 4, 5]
    ax = fig.add_subplot(6,3,sp+1)              #Plot positions = [3, 6, 9, 12, 15, 18]
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_index]], c=cb_orange, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[cat_index]], c=cb_dark_blue, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[cat_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    #Create labels for 'Men' and 'Women' in the charts    

    if cat_index == 0:
        ax.text(2005, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif cat_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2005, 35, 'Women')
       
#Show the charts
plt.show()
```

One interesting trend we can notice is that are that there are a number of majors with little to no gender gap, such as Architecture, Business, Agriculture, Social Sciences and History. With other majors such as Engineering, Health Professions, and Computer Science, there is a large gender gap and no indication if this gap will increase or decrease in the near future.

# Adding a horizontal line

Our charts are almost finished. For our last step, let's generate a horizontal line at y=50 on our charts. This will help us clearly see where the gender gap is between majors. For the line color, we will use the Color Blind 10 palette.
